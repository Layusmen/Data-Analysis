### Import Dependencies 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Import Data

```python
df = pd.read_csv('https://data.usaid.gov/api/views/dt8c-833c/rows.csv')
df
```

### Change the heading

```python
df.columns = ['GuaranteeNumber','TransactionReportID','GuaranteeCountryName','Amount','CurrencyName','DisbursementDate','EndDate',
           'Business Sector','CityTown','StateName','StateCode','StateCountryName','RegionName','IsWomanOwned?',
           'IsFirstTimeBorrower','BusinessSize','Latitude','Longitude']
df
```

```python
df.hist(figsize =(20,20))
```

```python
df.count()
```

```python
country_grp = df.groupby(['GuaranteeCountryName'])
country_grp.get_group('Nigeria')
```

```python
country_count = df['GuaranteeCountryName'].value_counts()
country_count.head(30)
```

### Top 20 Recipients of USAID Loans Among Countries

```python
country_grp = df.groupby(['GuaranteeCountryName'])
country_grp['Amount'].sum().sort_values(ascending=False).head(20).plot.bar(figsize=(15,15))
```

#### 20 Lowest Recipients of USAID DCA Loan

```python
country_grp['Amount'].sum().sort_values(ascending=True).head(20).plot.bar(figsize=(15,15))
```

### Loan Recipients Based on Sector

```python
business_grp = df.groupby(['Business Sector'])
business_grp['Amount'].sum().sort_values(ascending=False).head(20).plot.pie(figsize=(20,20))
```

### Note: 
* The above Chart does not capture the *NAN* values in the Business Sector column.
* So, we replace NAN with the word *Unknown*. Then plot the chart again.



```python
df['Business Sector'] = df['Business Sector'].fillna('Unknown')
```

```python
business_grp2= df.groupby(['Business Sector'])
business_grp2['Amount'].sum().sort_values(ascending=False).head(20).plot.pie(figsize=(20,20))
```

### First Time Borrower by Country 

```python
country_grp['IsFirstTimeBorrower'].sum().sort_values(ascending=False).head(20).plot.bar(figsize=(15,15))
```

### Business Owned By Women in Every Country

```python
country_grp['IsWomanOwned?'].sum().sort_values(ascending=False).head(20).plot.pie(figsize=(15,15))
```
